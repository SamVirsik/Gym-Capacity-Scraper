```python
url = "https://recwell.berkeley.edu/facilities/recreational-sports-facility-rsf/rsf-weight-room-crowd-meter/"  # Replace with the website URL you want to scrape
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
primary_div = soup.find_all('div', id = 'primary')[0]
main_div = primary_div.find('main')
ps = main_div.find_all('p')
newlink = ps[5].find("iframe").get('src')
newlink
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

driver.get(newlink)

try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.styles_waitTimeFullnessWrapper__3PRdQ"))
    )
except Exception as e:
    print("Error waiting for element:", e)
    driver.quit()
    exit()

soup = BeautifulSoup(driver.page_source, "html.parser")
print(soup.prettify())  # Or parse your specific data

driver.quit()
```
