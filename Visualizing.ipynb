```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'RSF_Data.csv' 
data = pd.read_csv(file_path, header=None, names=['Timestamp', 'Capacity'])

data['Timestamp'] = pd.to_datetime(data['Timestamp'])

data['Date'] = data['Timestamp'].dt.date

selected_date = pd.to_datetime('2025-01-31').date()
day_data = data[data['Date'] == selected_date]

plt.figure(figsize=(10, 5))
plt.plot(day_data['Timestamp'], day_data['Capacity'], marker='o', linestyle='-')
plt.xlabel('Time')
plt.ylabel('Gym Capacity')
plt.title(f'Gym Capacity on {selected_date}')
plt.xticks(rotation=45)
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(day_data['Capacity'], bins=range(day_data['Capacity'].min(), day_data['Capacity'].max() + 2), edgecolor='black', alpha=0.7)
plt.xlabel('Gym Capacity')
plt.ylabel('Frequency')
plt.title(f'Distribution of Gym Capacity on {selected_date}')
plt.grid(axis='y')
plt.show()
```
